```python
sc_stv1819['stv_total']=sc_stv1819['forward_trip_count']+sc_stv1819['reverse_trip_count']
sc_stv1819['stv_com_total']=sc_stv1819['forward_commute_trip_count']+sc_stv1819['reverse_commute_trip_count']
sc_stv1819['stv_rec_total']=sc_stv1819['forward_leisure_trip_count']+sc_stv1819['reverse_leisure_trip_count']
sc_stv1819['stv_com_ratio']=sc_stv1819['stv_com_total']/sc_stv1819['stv_total']
sc_stv1819['stv_rec_ratio']=sc_stv1819['stv_rec_total']/sc_stv1819['stv_total']
sc_stv1819.head()
```
